```python
import pygame
import random
import time
import json
from enum import Enum
from IPython.display import clear_output

# COLORES Y CONSTANTES
COLORES = {
    'negro': (10, 10, 20),
    'blanco': (255, 255, 255),
    'rojo': (220, 53, 69),
    'verde': (40, 167, 69),
    'azul': (0, 123, 255),
    'amarillo': (255, 193, 7),
    'naranja': (255, 140, 0),
    'gris': (108, 117, 125)
}

class Estado(Enum):
    NORMAL = 1
    ENVENENADO = 2
    QUEMADO = 3
    CONGELADO = 4

print('✅ Sistema inicializado')
```

```python
# SISTEMA DE HABILIDADES
class HabilidadEspecial:
    def __init__(self, nombre, descripcion, daño, costo_energia, cooldown=0):
        self.nombre = nombre
        self.descripcion = descripcion
        self.daño = daño
        self.costo_energia = costo_energia
        self.cooldown = cooldown
        self.cooldown_actual = 0
    
    def puede_usarse(self):
        return self.cooldown_actual <= 0
    
    def usar(self):
        self.cooldown_actual = self.cooldown

# SISTEMA DE ITEMS
class Item:
    def __init__(self, nombre, tipo, efecto, cantidad=1):
        self.nombre = nombre
        self.tipo = tipo
        self.efecto = efecto
        self.cantidad = cantidad

# PERSONAJES AVANZADOS
class PersonajeAvanzado:
    def __init__(self, nombre, vida, ataque, defensa, velocidad, emoji):
        self.nombre = nombre
        self.vida_max = vida
        self.vida = vida
        self.ataque_base = ataque
        self.ataque = ataque
        self.defensa_base = defensa
        self.defensa = defensa
        self.velocidad = velocidad
        self.emoji = emoji
        self.nivel = 1
        self.experiencia = 0
        self.energia = 100
        self.estado = Estado.NORMAL
        self.defensa_activa = False
        
        # Habilidades especiales
        self.habilidades = [
            HabilidadEspecial('Rasengan', 'Ataque de poder extremo', 45, 40, 2),
            HabilidadEspecial('Substitución', 'Evita ataques próximos', 0, 20, 3),
            HabilidadEspecial('Regeneración', 'Restaura 50 HP', 0, 30, 2)
        ]
        
        # Inventario
        self.inventario = [
            Item('Pastilla Energética', 'energia', 50, 3),
            Item('Antídoto', 'efecto', 'estado', 2),
            Item('Elixir de Vida', 'curacion', 100, 1)
        ]
    
    def atacar_normal(self, enemigo):
        daño_base = self.ataque + random.randint(-3, 8)
        daño_final = max(daño_base - (enemigo.defensa // 3), 1)
        
        if enemigo.defensa_activa:
            daño_final = int(daño_final * 0.6)
        
        enemigo.vida -= daño_final
        self.energia = min(self.energia + 10, 100)
        
        return daño_final
    
    def defender(self):
        self.defensa_activa = True
        self.defensa = int(self.defensa_base * 2.5)
        return f'🛡️ {self.nombre} se prepara para defender!'
    
    def curar(self, cantidad):
        cantidad_real = min(cantidad, self.vida_max - self.vida)
        self.vida += cantidad_real
        return f'💚 {self.nombre} recupera {cantidad_real} HP!'
    
    def esta_vivo(self):
        return self.vida > 0
    
    def mostrar_barra_vida(self):
        porcentaje = max(0, (self.vida / self.vida_max) * 20)
        barra = '█' * int(porcentaje) + '░' * (20 - int(porcentaje))
        
        if self.vida > self.vida_max * 0.6:
            color = '🟢'
        elif self.vida > self.vida_max * 0.3:
            color = '🟡'
        else:
            color = '🔴'
        
        return f'{color} {self.nombre} Lv{self.nivel}: [{barra}] {self.vida}/{self.vida_max}'

print('✅ Clases de personajes creadas')
```

```python
# SISTEMA DE BATALLA AVANZADO
class SistemaBatallaAvanzada:
    def __init__(self, equipo_jugador, equipo_enemigos):
        self.equipo_jugador = equipo_jugador
        self.equipo_enemigos = equipo_enemigos
        self.turno = 0
    
    def mostrar_interfaz(self):
        clear_output(wait=True)
        print('\n' + '='*70)
        print(f'⚔️ TURNO {self.turno} ⚔️'.center(70))
        print('='*70 + '\n')
        
        print('🟢 EQUIPO JUGADOR:')
        for p in self.equipo_jugador:
            print(f'  {p.mostrar_barra_vida()}')
        
        print('\n' + '-'*70 + '\n')
        
        print('🔴 ENEMIGOS:')
        for e in self.equipo_enemigos:
            print(f'  {e.mostrar_barra_vida()}')
        
        print('\n' + '='*70)
    
    def batalla_completa(self):
        print('🎬 ¡COMIENZA LA BATALLA! 🎬\n')
        time.sleep(1)
        
        while any(p.esta_vivo() for p in self.equipo_jugador) and any(e.esta_vivo() for e in self.equipo_enemigos):
            self.turno += 1
            self.mostrar_interfaz()
            
            # Turno del jugador
            for personaje in self.equipo_jugador:
                if not personaje.esta_vivo():
                    continue
                
                enemigo = random.choice([e for e in self.equipo_enemigos if e.esta_vivo()])
                accion = random.choice(['atacar', 'atacar', 'defenderse'])
                
                if accion == 'atacar':
                    daño = personaje.atacar_normal(enemigo)
                    print(f'⚡ {personaje.nombre} ataca a {enemigo.nombre} por {daño} de daño')
                else:
                    print(personaje.defender())
            
            time.sleep(0.5)
            
            # Turno del enemigo
            for enemigo in self.equipo_enemigos:
                if not enemigo.esta_vivo():
                    continue
                
                personaje = random.choice([p for p in self.equipo_jugador if p.esta_vivo()])
                daño = enemigo.atacar_normal(personaje)
                print(f'💢 {enemigo.nombre} ataca a {personaje.nombre} por {daño} de daño')
            
            time.sleep(0.5)
        
        # Resultado
        clear_output(wait=True)
        print('\n' + '='*70)
        
        if any(p.esta_vivo() for p in self.equipo_jugador):
            print('🏆 ¡VICTORIA! ¡HAS GANADO LA BATALLA! 🏆'.center(70))
        else:
            print('💀 DERROTA... Los enemigos fueron demasiado fuertes 💀'.center(70))
        
        print('='*70 + '\n')

print('✅ Sistema de batalla creado')
```

```python
# CREAR PERSONAJES
naruto = PersonajeAvanzado('Naruto', 120, 22, 10, 12, '🧡')
sasuke = PersonajeAvanzado('Sasuke', 115, 25, 12, 14, '🖤')
kakashi = PersonajeAvanzado('Kakashi', 100, 20, 15, 16, '👁️')

zabuza = PersonajeAvanzado('Zabuza', 100, 20, 12, 10, '⚔️')
haku = PersonajeAvanzado('Haku', 90, 18, 10, 13, '❄️')

print('✅ Personajes creados')
print(f'\n⭐ Equipo Jugador:')
for p in [naruto, sasuke, kakashi]:
    print(f'   {p.emoji} {p.nombre} - {p.vida} HP')

print(f'\n⭐ Enemigos:')
for e in [zabuza, haku]:
    print(f'   {e.emoji} {e.nombre} - {e.vida} HP')
```

```python
# ⚔️ EJECUTAR BATALLA
batalla = SistemaBatallaAvanzada([naruto, sasuke, kakashi], [zabuza, haku])
batalla.batalla_completa()
```
